```python
# --- Imports ---
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import math
```

```python
# --- Load tables ---
manifest = pd.read_csv(CSV_MANIFEST)
exploded = pd.read_csv(CSV_EXPLODED)

for col in ["csa","indicator_token","period"]:
    if col in exploded.columns:
        exploded[col] = exploded[col].astype(str)

CSAS = sorted(exploded["csa"].unique())
INDICATORS = sorted(exploded["indicator_token"].unique())
PERIODS = sorted(exploded["period"].unique())

print("CSAs:", CSAS)
print("Indicators:", INDICATORS)
print("Periods:", PERIODS)
```

```python
# Select indicator & period
INDICATOR_SELECT = "SU"       # e.g., "HD", "WSDI", "TXX", "RX5DAY", "CDD"
PERIOD_SELECT    = "2036-2065" # e.g., "1981-2010", "1991-2020", "2036-2065"

subset = exploded[(exploded["indicator_token"]==INDICATOR_SELECT) &
                  (exploded["period"]==PERIOD_SELECT)].copy()

# One row per CSA (pick a corresponding manifest row; prefer local files)
rows = []
for csa in sorted(subset["csa"].unique()):
    cand = subset[subset["csa"]==csa][["basename","relpath"]].drop_duplicates()
    msel = manifest.merge(cand, on=["basename","relpath"], how="inner")
    if msel.empty:
        continue
    if (msel["storage"]=="file").any():
        r = msel[msel["storage"]=="file"].iloc[0].to_dict()
        r["fullpath"] = str((Path(GITHUB_DIR) / r["relpath"]).resolve())
        r["open_kind"] = "file"
    else:
        r = msel.iloc[0].to_dict()
        r["fullpath"] = ""
        r["open_kind"] = "zip"
    r["csa"] = csa
    rows.append(r)

panel_df = pd.DataFrame(rows)
print("Panel rows:", len(panel_df))
panel_df.head(10)
```

```python
import numpy as np

def to_numeric_days(da):
    """If DataArray is timedelta64, convert to float days and set units."""
    if np.issubdtype(da.dtype, np.timedelta64):
        da = (da / np.timedelta64(1, "D")).astype("float64")
        # keep any existing attrs, just adjust units nicely
        units = str(da.attrs.get("units", "")).strip()
        da.attrs["units"] = "days" if units == "" else f"{units} (days)"
    return da
```

```python
def choose_var(ds, indicator_token: str):
    candidates = list(ds.data_vars)
    hits = [v for v in candidates if indicator_token.lower() in v.lower()]
    return hits[0] if hits else candidates[0]

def time_subset_for_period(ds_or_da, period_label: str):
    if "time" not in getattr(ds_or_da, "coords", {}) and "time" not in getattr(ds_or_da, "dims", []):
        return ds_or_da
    try:
        y1, y2 = map(int, period_label.split("-"))
        import pandas as pd
        years = pd.DatetimeIndex(ds_or_da["time"].values).year
        mask = (years >= y1) & (years <= y2)
        return ds_or_da.sel(time=mask)
    except Exception:
        return ds_or_da

p2_all, p98_all = [], []
snapshots = {}
skipped = []

for _, r in panel_df.iterrows():
    if r["open_kind"] != "file":
        skipped.append((r["csa"], "only in zip"))
        continue
    try:
        ds = xr.open_dataset(r["fullpath"], engine="netcdf4", decode_timedelta=True)
        var = choose_var(ds, INDICATOR_SELECT)
        da = ds[var].squeeze()
        da = time_subset_for_period(da, PERIOD_SELECT)
        if "time" in getattr(da, "dims", []):
            da_plot = da.mean("time")
        else:
            da_plot = da
        arr = np.asarray(da_plot.values, dtype="float64")
        arr = arr[np.isfinite(arr)]
        if arr.size == 0:
            skipped.append((r["csa"], "no finite values"))
        else:
            p2_all.append(np.nanpercentile(arr, 2))
            p98_all.append(np.nanpercentile(arr, 98))
            snapshots[r["csa"]] = (da_plot, ds)
    except Exception as e:
        skipped.append((r["csa"], f"error: {type(e).__name__}"))

# reduce time for map
if "time" in getattr(da, "dims", []):
    da_plot = da.mean("time")
else:
    da_plot = da

# NEW: normalize timedelta → float days
da_plot = to_numeric_days(da_plot)

arr = np.asarray(da_plot.values, dtype="float64")
arr = arr[np.isfinite(arr)]


if p2_all and p98_all:
    vmin = float(np.nanmin(p2_all))
    vmax = float(np.nanmax(p98_all))
    if not np.isfinite(vmin) or not np.isfinite(vmax) or vmax <= vmin:
        vmin, vmax = None, None
else:
    vmin, vmax = None, None

print("Common scale:", vmin, "→", vmax)
print("Skipped:", skipped[:5], "… total", len(skipped))
```

```python
# --- Helpers for dtype + period ---
import numpy as np
import pandas as pd
import xarray as xr

def choose_var(ds, indicator_token: str):
    """Prefer variable containing token (case-insensitive); else first var."""
    candidates = list(ds.data_vars)
    hits = [v for v in candidates if indicator_token.lower() in v.lower()]
    return hits[0] if hits else candidates[0]

def time_subset_for_period(da: xr.DataArray, period_label: str) -> xr.DataArray:
    """If time axis exists, subset to YYYY-YYYY."""
    if "time" not in getattr(da, "dims", []):
        return da
    try:
        y1, y2 = map(int, period_label.split("-"))
        years = pd.DatetimeIndex(da["time"].values).year
        return da.sel(time=(years >= y1) & (years <= y2))
    except Exception:
        return da

def to_numeric_days_if_timedelta(da: xr.DataArray) -> xr.DataArray:
    """timedelta64 → float days; else force float64 for robust plotting."""
    if np.issubdtype(da.dtype, np.timedelta64):
        out = (da / np.timedelta64(1, "D")).astype("float64")
        units = str(da.attrs.get("units", "")).strip()
        out.attrs["units"] = "days" if units == "" else f"{units} (days)"
        return out
    return da.astype("float64", copy=False)
```

```python
# --- Compute common color scale (robust percentiles) and collect snapshots ---
p2_all, p98_all = [], []
snapshots = {}   # {csa: (numeric_da_plot, open_dataset_handle)}
skipped = []

for _, r in panel_df.iterrows():
    if r.get("open_kind") != "file":
        skipped.append((r["csa"], "only in zip (not opened)"))
        continue

    try:
        ds = xr.open_dataset(r["fullpath"], engine="netcdf4", decode_timedelta=True)
        var = choose_var(ds, INDICATOR_SELECT)
        da = ds[var].squeeze()

        # Subset period if applicable
        da = time_subset_for_period(da, PERIOD_SELECT)

        # Reduce to a single map
        if "time" in getattr(da, "dims", []):
            with xr.set_options(keep_attrs=True):
                da_plot = da.mean("time", skipna=True)
        else:
            da_plot = da

        # Normalize dtype (timedelta → days; ensure float64)
        da_plot = to_numeric_days_if_timedelta(da_plot)

        arr = np.asarray(da_plot.values, dtype="float64")
        arr = arr[np.isfinite(arr)]
        if arr.size == 0:
            skipped.append((r["csa"], "no finite values"))
            try: ds.close()
            except Exception: pass
            continue

        p2_all.append(np.nanpercentile(arr, 2))
        p98_all.append(np.nanpercentile(arr, 98))
        snapshots[r["csa"]] = (da_plot, ds)

    except Exception as e:
        skipped.append((r["csa"], f"error: {type(e).__name__}: {e}"))

# Decide shared scale
if p2_all and p98_all:
    vmin = float(np.nanmin(p2_all))
    vmax = float(np.nanmax(p98_all))
    if not np.isfinite(vmin) or not np.isfinite(vmax) or vmax <= vmin:
        vmin, vmax = None, None
else:
    vmin, vmax = None, None

print("Common scale:", vmin, "→", vmax)
print("Skipped (first 5):", skipped[:5], f"… total {len(skipped)}")
```

```python
# --- Render and save panel (constrained layout, no tight_layout) ---
N = len(snapshots)
if N == 0:
    print("No CSA snapshots available to plot.")
else:
    ncols = 3 if N >= 6 else 2 if N >= 3 else 1
    nrows = int(np.ceil(N / ncols))

    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=(5*ncols, 4.5*nrows),
        constrained_layout=True  # <- key change
    )
    axes = np.atleast_1d(axes).reshape(nrows, ncols)

    mappable = None
    order = sorted(snapshots.keys())

    k = 0
    for i in range(nrows):
        for j in range(ncols):
            ax = axes[i, j]
            if k >= N:
                ax.set_visible(False)
                continue
            csa = order[k]; k += 1
            da_plot, ds = snapshots[csa]
            im = da_plot.plot(ax=ax, add_colorbar=False, vmin=vmin, vmax=vmax)
            ax.set_title(csa)
            if mappable is None:
                mappable = im

    # Shared colorbar (works with constrained_layout)
    if mappable is not None:
        cbar = fig.colorbar(mappable, ax=axes, fraction=0.025, pad=0.02)
        units = str(da_plot.attrs.get("units", "")) if "da_plot" in locals() else ""
        cbar.set_label(units)

    fig.suptitle(f"{INDICATOR_SELECT} — {PERIOD_SELECT}")
    out_png = OUT_DIR / f"panel_{INDICATOR_SELECT}_{PERIOD_SELECT}.png"
    plt.savefig(out_png, dpi=200)  # optionally add bbox_inches='tight'
    plt.show()
    print("Saved panel to:", out_png)

    for (da_plot, ds) in snapshots.values():
        try: ds.close()
        except Exception: pass
```
